```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('/Users/dongyokim/Library/Mobile Documents/com~apple~CloudDocs/2022:1/비즈니스 애널리틱스/실습/판별분석_실습_9주차 .xlsx', sheet_name='Data')
```

```python
df
```

```python
df_data = df[['Income', 'Lot_Size']]
df_target = df['Ownership']
```

```python
# 데이터 정규 스케일링
df_scaled_data = StandardScaler().fit_transform(df_data)
```

```python
df_scaled_data
```

```python
# 2개의 클래스로 구분하기 위한 LDA 생성
lda = LinearDiscriminantAnalysis(n_components=1)
```

```python
# fit()호출 시 target값 입력 
lda.fit(df_scaled_data, df_target)
df_lda = lda.transform(df_scaled_data)
```

```python
df_lda
```

```python
from sklearn.metrics import accuracy_score
pred = lda.predict(df_scaled_data)
print(accuracy_score(df_target, pred))
```

```python
df = pd.read_excel('/Users/dongyokim/Library/Mobile Documents/com~apple~CloudDocs/2022:1/비즈니스 애널리틱스/Data/Accidents.xlsx', sheet_name='Data')
```

```python
df = pd.concat([df, pd.get_dummies(df['LGTCON_I_R'], prefix='LGTCON_I_R')],axis=1)
```

```python
df.drop(columns='LGTCON_I_R', inplace=True)
```

```python
df = pd.concat([df, pd.get_dummies(df['SUR_COND'], prefix='SUR_COND')],axis=1)
```

```python
df.drop(columns='SUR_COND', inplace=True)
```

```python
df = pd.concat([df, pd.get_dummies(df['TRAF_WAY'], prefix='TRAF_WAY')],axis=1)
```

```python
df.drop(columns='TRAF_WAY', inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(df.drop(columns='MAX_SEV_IR') , df['MAX_SEV_IR'], test_size=0.4, random_state=0) 
```

```python
lda = LinearDiscriminantAnalysis(n_components=2)

lda.fit(X_train, y_train)

train_pred = lda.predict(X_train)

print('학습 데이터의 정확도: {0:.4f}'.format(accuracy_score(y_train, train_pred)))
print('학습 데이터의 confusion matrix :\n {}'.format(confusion_matrix(y_train, train_pred)))
```

```python
test_pred = lda.predict(X_test)
print('검증 데이터의 정확도: {0:.4f}'. format(accuracy_score(y_test, test_pred)))
print('학습 데이터의 confusion matrix :\n {}'.format(confusion_matrix(y_test, test_pred)))
```
