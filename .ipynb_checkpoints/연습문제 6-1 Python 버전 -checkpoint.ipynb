```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
import missingno as msno
import statsmodels.api as sm
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

# 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
matplotlib.rcParams['axes.unicode_minus'] = False

# 한글 폰트 설정
plt.rcParams['font.family'] = 'AppleGothic'
```

```python
# 데이터 불러오기 
data = pd.read_excel('BostonHousing.xlsx', sheet_name = 'Data')
data.head()
```

## 1. A ~ C번 풀이

```python
# A번 풀이 
x = data[['CRIM', 'CHAS', 'RM']]
# 절편 값 추가 
x1 = sm.add_constant(x, has_constant='add')
target = data['MEDV']

lr_model = sm.OLS(target,x1)
fitted_lr_model=lr_model.fit()

fitted_lr_model.summary()
```

```python
# 회귀계수 도출 
fitted_lr_model.params
```

```python
print('절편 값:', fitted_lr_model.params[0])
print('회귀 계수값:', [fitted_lr_model.params[1], fitted_lr_model.params[2], fitted_lr_model.params[3]])
```

```python
print('회귀식 : {}x1(CRIM) + {}x2(CHAS) + {}x3(RM) + {}(절편)'.format(fitted_lr_model.params[1], fitted_lr_model.params[2], fitted_lr_model.params[3], fitted_lr_model.params[0]))
```

```python
# 찰스강 경계에 위치하지 않고, 범죄율이 0.1, 평균방의 개수가 6개인 보스턴 주택단지의 주택가격 중앙값 예측 값
# 리스트 첫 번째 넣은 값은 const(상수) 값
pred = fitted_lr_model.predict([1, 0.1, 0, 6])

print('찰스강 경계에 위치하지 않고, 범죄율이 0.1, 평균방의 개수가 6개인 보스턴 주택단지의 주택가격 중앙값 예측 값 : {0:.3f}'.format(pred[0]))
```

```python
# 찰스강 경계에 위치하지 않고, 범죄율이 0.1, 평균방의 개수가 6개인 보스턴 주택단지의 주택가격 중앙값 실제값 33.1 
data[data['CRIM'] == 0.1]
```

```python
value = data[data['CRIM'] == 0.1]['MEDV'].values[0]
print('실제값 =', value)
print('예측값 = {0:.3f}'.format(pred[0]) )
print('예측오차 = {0:.3f}'.format(value - pred[0]))
```

```python
## 변수별 산점도 시각화
sns.pairplot(data)
plt.show()
```

```python
# 모델링 
x = data.drop(['MEDV'], axis=1)
x1 = sm.add_constant(x, has_constant='add')
target = data['MEDV']
lr_model = sm.OLS(target,x1)
fitted_lr_model=lr_model.fit()

fitted_lr_model.summary()
```

```python
# VIF를 이용한 다중공선성 진단 
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(
    x.values, i) for i in range(x.shape[1])]
vif["features"] = x.columns

vif
```

```python
# 데이터 분할 
from sklearn.model_selection import train_test_split
# 상관관계가 강하고 VIF 값이 높은 변수들 제거 
X = data.drop(['MEDV', 'RM', 'NOX', 'TAX', 'PTRATIO', 'INDUS', 'RAD', 'CAT. MEDV'], axis=1)
X = sm.add_constant(X, has_constant='add')
y = data['MEDV']
train_x, test_x, train_y, test_y = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=0)
print(train_x.shape, test_x.shape, train_y.shape, test_y.shape)
```

```python
# Train the MLR / 회귀모델적합
full_model = sm.OLS(train_y, train_x)
fitted_full_model = full_model.fit()
```

```python
fitted_full_model.summary() ## R2가 높고, 대부분의 변수들이 유의
```
